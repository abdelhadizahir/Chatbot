```python
from nltk.chat.util import Chat, reflections
pairs = [
    [
        r"je m'appelle (.*)",
        ["Bonjour %1, comment allez vous ?",]
    ],
     [
        r"quel est votre nom ?",
        ["Je m'appelle Chatty et je suis un chatbot ?",]
    ],
    [
        r"comment ça va ?",
        ["e vais bien\net toi ?",]
    ],
    [
        r"sorry (.*)",
        ["C'est bon, ne t'en fais pas",]
    ],
    [
        r"je vais bien",
        ["Ravi d'entendre ça "," D'accord :)",]
    ],
    [
        r"salut|bonsoir|bonjour",
        ["Bonjour "," Salut",]
    ],
    [
        r"(.*) age?",
        ["I'm a computer program dude\nSeriously you are asking me this?",]
        
    ],
    [
        r"what (.*) want ?",
        ["Make me an offer I can't refuse",]
        
    ],
    [
        r"(.*) created ?",
        ["Nagesh created me using Python's NLTK library ","top secret ;)",]
    ],
    [
        r"(.*) (location|city) ?",
        ['Chennai, Tamil Nadu',]
    ],
    [
        r"how is weather in (.*)?",
        ["Weather in %1 is awesome like always","Too hot man here in %1","Too cold man here in %1","Never even heard about %1"]
    ],
    [
        r"i work in (.*)?",
        ["%1 is an Amazing company, I have heard about it. But they are in huge loss these days.",]
    ],
[
        r"(.*)raining in (.*)",
        ["No rain since last week here in %2","Damn its raining too much here in %2"]
    ],
    [
        r"how (.*) health(.*)",
        ["I'm a computer program, so I'm always healthy ",]
    ],
    [
        r"(.*) (sports|game) ?",
        ["I'm a very big fan of Football",]
    ],
    [
        r"who (.*) sportsperson ?",
        ["Messy","Ronaldo","Roony"]
],
    [
        r"who (.*) (moviestar|actor)?",
        ["Brad Pitt"]
],
    [
        r"quit",
        ["BBye take care. See you soon :) ","It was nice talking to you. See you soon :)"]
],
]
def chatty():
    print("Bonjour, bienvenue dans notre ChatBot ;)\nVeuillez saisir en minuscule pour démarrer une conversation. Tapez quit pour quitter ") #default message at the start
    chat = Chat(pairs, reflections)
    chat.converse()

    

chatty()
```
